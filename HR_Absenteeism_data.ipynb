## Task 3 - Visualizing the Data

```python
data
```

1.Histogram

```python
plt.figure(figsize=(10,6))
sns.histplot(x='Age',data=data,kde=True,color='blue',bins=100)
plt.title('Age - Histogram')
plt.vlines(mean_age,ymin=0,ymax=300,color='r',label='Mean')
plt.legend()
plt.show()
```

2.Bargraph

```python
plt.figure(figsize=(10,8))
sns.barplot(y="Age",x='Gender',data=data,palette="Set1")
labels=['Female','Male']
#plt.legend(labels=labels)
plt.title('Bargraph of Gender vs Age')
plt.show()
```

### 3.Count plot
- Based on cities separated by Gender

```python
plt.figure(figsize=(15,8))
sns.countplot(x='City',data=data[0:10],hue='Gender',palette="Set1")
plt.title('Count of People in Different cities')
plt.show()
```

### 4.Scatter plot 
- To observe the scattering of different values 
- The values are taken upto 150 samples

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='Age',y='JobTitle',data=data[0:150])
plt.title('Scatterplot of Age vs JobTitle')
plt.show()
```

### 5.Heatmap
- Correlation between values (Which are Numericals only)

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(),annot=True,linewidth=1)
plt.show()
```
